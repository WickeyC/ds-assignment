```python
gold_price = pd.read_csv('processed_price_gold.csv',na_values=['null'],index_col='Date',parse_dates=True,infer_datetime_format=True)
```

## 6. Further Analysis

### 6.1 Daily Return Rate

```python
def compute_daily_returns(df):
    daily_return = (df / df.shift(1)) - 1
    daily_return[0] = 0
    return daily_return
```

```python
GLD_ajclose = gold_price['GOLD_ajclose']
SP_ajclose = gold_price['SP_ajclose']
DJ_ajclose  = gold_price['DJ_ajclose']
EG_ajclose =  gold_price['EG_ajclose']
USO_ajclose = gold_price['USO_ajclose']
GDX_ajclose = gold_price['GDX_ajclose']
EU_price      = gold_price['EU_price']
OF_price      = gold_price['OF_price']
OS_price      = gold_price['OS_price']
SF_price      = gold_price['SF_price']
USB_price      = gold_price['USB_price']
PLT_price      = gold_price['PLT_price']
PLD_price      = gold_price['PLD_price']
RHO_price      = gold_price['RHO_price']
USDI_price      = gold_price['USDI_price']

GLD_daily_return = compute_daily_returns(GLD_ajclose)
SP_daily_return = compute_daily_returns(SP_ajclose)
DJ_adj_return    = compute_daily_returns(DJ_ajclose)
EG_adj_return     = compute_daily_returns(EG_ajclose)
USO_adj_return    = compute_daily_returns(USO_ajclose)
GDX_adj_return   =compute_daily_returns(GDX_ajclose)
EU_return        = compute_daily_returns(EU_price)
OF_price         =compute_daily_returns(OF_price)
OS_price         =compute_daily_returns(OS_price)
SF_price         =compute_daily_returns(SF_price)
USB_price         =compute_daily_returns(USB_price)
PLT_price         =compute_daily_returns(PLT_price)
PLD_price         =compute_daily_returns(PLD_price)
rho_price         =compute_daily_returns(RHO_price)
USDI_price         =compute_daily_returns(USDI_price)
```

```python
df_d = pd.DataFrame({'GLD':GLD_daily_return, 'EG':EG_adj_return, 'GDX':GDX_adj_return})

daily_ax = df_d[-100:].plot(title='Last 100 records of daily return of Gold Related Features',figsize=(15,8))

daily_ax.set_ylabel('Daily return')
daily_ax.legend(loc='lower left')
plt.show()
```

```python
df_d = pd.DataFrame({'SF':SF_price, 'PLT':PLT_price, 'PLD':PLD_price, 'RHO':rho_price})

daily_ax = df_d[-100:].plot(title='Last 100 records of daily return of Precious Materials',figsize=(15,8))

daily_ax.set_ylabel('Daily return')
daily_ax.legend(loc='lower left')
plt.show()
```
